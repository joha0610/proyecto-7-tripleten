```python
import pandas as pd

# Cargar el archivo CSV con la ruta correcta
car_data = pd.read_csv("vehicles_us.csv")  

# Mostrar las primeras filas para verificar que se cargó bien
car_data.head()
```

```python
fig = px.histogram(car_data, x="odometer")  # Histograma del odómetro
fig.show()

```

```python
fig = px.scatter(car_data, x="odometer", y="price")  # Relación entre odómetro y precio
fig.show()
```

```python
fig = px.bar(car_data, x="model", y="price")  # Precio promedio por modelo
fig.show()

```
